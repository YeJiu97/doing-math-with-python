# 笛卡尔坐标系

笛卡尔坐标系就是直角坐标系和斜角坐标系的统称。 相交于原点的两条数轴，构成了平面仿射坐标系。如两条数轴上的度量单位相等，则称此仿射坐标系为笛卡尔坐标系。两条数轴互相垂直的笛卡尔坐标系，称为笛卡尔直角坐标系，否则称为笛卡尔斜角坐标系。需要指出的是，请将数学中的笛卡尔坐标系与电影《异次元杀阵》中的笛卡尔坐标相区分，电影中的定义与数学中定义有出入，请勿混淆。

二维的直角坐标系是由两条相互垂直、0 点重合的数轴构成的。在平面内，任何一点的坐标是根据数轴上对应的点的坐标设定的。在平面内，任何一点与坐标的对应关系，类似于数轴上点与坐标的对应关系。采用直角坐标，几何形状可以用代数公式明确的表达出来。几何形状的每一个点的直角坐标必须遵守这代数公式。

**元组**

元组，又称为Tuple，是一种不可变的序列类型。它的特点是包含不同的元素，并且这些元素不能被修改。在Python编程语言中，元组使用小括号和逗号隔开元素。尽管元素无法更改，但元组支持许多操作，包括索引、拼接和成员检查等。

此外，元组在关系数据库中也有其特殊含义。在那里，元组是关系的基本概念，关系即一张表，表中的每行（即数据库中的每条记录）就是一个元组，每列就是一个属性。

```python
tuple_01 = (1, 2, 3, 4, 5)

type(tuple_01)
```

```python
tuple_01[0]
```

```python
try:
    tuple_01[0] = 'testing'  # 测试一下
except TypeError:
    print("Tuple不支持修改")
```

# 用公式绘图

**牛顿万有引力公式**

牛顿的万有引力公式为：

$F = G \frac{m_1 m_2}{r^2}$

```python
import matplotlib.pyplot as plt

# 首先是绘制图形函数
def draw_graph(x, y):
    plt.plot(x, y, marker = 'o')
    plt.xlabel('Distance in meters')
    plt.ylabel('Gravitational force in newtons')
    plt.title('Gravitational force and distance')
    plt.show()

# 接着是万有引力
def generate_F_r():
    # Generate value for r
    r = range(100, 1001, 50)
    F = []
    # Constant, G
    G = 6.674*(10**-11)
    # 两个物体的质量
    mass1 = float(input("请输入第一个物体的质量（单位：kg）："))
    mass2 = float(input("请输入第二个物体的质量（单位：kg）："))
    # calculate force and add it to the list, F
    for dist in r:
        force = G * (mass1 * mass2) / (dist**2)
        F.append(force)

    # call the draw_graph function
    draw_graph(r, F)

# main function
if __name__ == '__main__':
    generate_F_r()
```

**抛物运动**

```python
from matplotlib import pyplot as plt
import math

def draw_graph(x, y):
    plt.plot(x, y)
    plt.xlabel('X-coordinate')
    plt.ylabel('Y-coordinate')
    plt.title("Projectile motion of a ball")

def frange(start, final, interval):
    numbers = []
    while start < final:
        numbers.append(start)
        start = start + interval
    return numbers

def draw_trajectory(u, theta):
    theta = math.radians(theta)
    g = 9.8

    t_flight = 2 * u * math.sin(theta) / g
    intervals = frange(0, t_flight, 0.001)

    x = []
    y = []

    for t in intervals:
        x.append(u * math.cos(theta) * t)
        y.append(u * math.sin(theta) * t - 0.5*g*t*t)

    draw_graph(x, y)

if __name__ == '__main__':
    try:
        u = float(input('Enter the initial velocity (m/s): '))
        theta = float(input('Enter the angle of projection (degree): '))
    except ValueError:
        print('You entered an invalid input')
    else:
        draw_trajectory(u, theta)
        plt.show()
```

```python
from matplotlib import pyplot as plt
import math

def draw_graph(x, y):
    plt.plot(x, y)
    plt.xlabel('X-coordinate')
    plt.ylabel('Y-coordinate')
    plt.title("Projectile motion of a ball")

def frange(start, final, interval):
    numbers = []
    while start < final:
        numbers.append(start)
        start = start + interval
    return numbers

def draw_trajectory(u, theta):
    theta = math.radians(theta)
    g = 9.8

    t_flight = 2 * u * math.sin(theta) / g
    intervals = frange(0, t_flight, 0.001)

    x = []
    y = []

    for t in intervals:
        x.append(u * math.cos(theta) * t)
        y.append(u * math.sin(theta) * t - 0.5*g*t*t)

    draw_graph(x, y)

if __name__ == '__main__':
    try:
        m_list = [float(input("请输入物体{}的质量（kg）：".format(i+1))) for i in range(3)]
        theta = float(input("请输入抛出角度（度）："))
    except ValueError:
        print('你输入了一个无效的输入')
    else:
        for i, m in enumerate(m_list):
            draw_trajectory(m, theta)
            plt.legend(['物体{}'.format(i+1)])
        plt.show()
```

**斐波那契按数列和黄金比例**

```python
import matplotlib.pyplot as plt

def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0, 1]
    else:
        fib = [0, 1]
        for i in range(2, n):
            fib.append(fib[-1] + fib[-2])
        return fib

def ratio_visualization(n):
    fib = [0, 1]
    ratios = []
    for i in range(2, n + 1):
        fib.append(fib[-1] + fib[-2])
        if i == 1:
            continue
        ratios.append(fib[i] / fib[i - 1])
    plt.plot(ratios)
    plt.xlabel('Index')
    plt.show()

n = int(input("请输入一个数字："))
ratio_visualization(n)
```

1. 生成这么多个斐波那契数字
2. 然后将这些数字的相邻的笔直添加到ratio的列表中
3. 然后进行可视化
