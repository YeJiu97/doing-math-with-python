# 使用matplotlib的patches绘制几何图形

在matplotlib中，patches可以用来绘制几何图形。

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 创建一个新的图形
fig, ax = plt.subplots()

# 创建一个矩形
rect = patches.Rectangle((0.1, 0.1), 0.3, 0.6, linewidth=2, edgecolor='r', facecolor='none')
ax.add_patch(rect)

# 创建一个圆形
circle = patches.Circle((0.7, 0.5), 0.2, color='b', alpha=0.5)
ax.add_patch(circle)

# 创建一个多边形
polygon = patches.RegularPolygon((0.5, 0.8), numVertices=5, radius=0.2, color='g', alpha=0.8)
ax.add_patch(polygon)

# 设置图形的范围
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

# 显示图形
plt.show()
```

绘制树叶：

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.path import Path
import numpy as np

# 定义树叶的路径
leaf_path_data = [
    (Path.MOVETO, [0, 0]),           # 移动到起始点
    (Path.CURVE4, [0.2, 0.2]),      # 二次贝塞尔曲线控制点
    (Path.CURVE4, [0.5, 0.5]),      # 二次贝塞尔曲线控制点
    (Path.CURVE4, [0.8, 0.2]),      # 二次贝塞尔曲线控制点
    (Path.LINETO, [0, 0]),           # 直线到起始点，形成闭合路径
]

leaf_path_codes, leaf_path_verts = zip(*leaf_path_data)
leaf_path = Path(leaf_path_verts, leaf_path_codes)

# 创建一个图形和坐标轴
fig, ax = plt.subplots()

# 创建一个树叶形状的PathPatch
leaf_patch = patches.PathPatch(leaf_path, facecolor='green', lw=2)
ax.add_patch(leaf_patch)

# 设置坐标轴范围
ax.set_xlim(-0.2, 1)
ax.set_ylim(-0.2, 0.6)

# 隐藏坐标轴
ax.axis('off')

# 显示图形
plt.show()
```

使用Dot Points的形式来绘制：

```python
import matplotlib.pyplot as plt
import numpy as np

# 生成蕨类植物的枝叶轮廓，使用随机生成的点
num_points = 1000

# 生成随机枝叶轮廓的坐标
theta = np.linspace(0, 2*np.pi, num_points)
r = np.sqrt(np.random.rand(num_points))  # 随机生成半径
x = r * np.cos(theta)
y = r * np.sin(theta)

# 可视化蕨类植物的枝叶轮廓
plt.figure(figsize=(8, 8))
plt.plot(x, y, 'g.')  # 使用绿色点来表示枝叶轮廓
plt.title('Fern Plant Outline')
plt.show()
```

绘制Sierpinkski三角：

```python
import matplotlib.pyplot as plt
import numpy as np

def sierpinski(ax, p1, p2, p3, depth):
    if depth > 0:
        # 计算三角形的中点
        mid_1_2 = (p1 + p2) / 2
        mid_2_3 = (p2 + p3) / 2
        mid_1_3 = (p1 + p3) / 2
        
        # 绘制当前三角形
        ax.plot([p1[0], p2[0], p3[0], p1[0]], [p1[1], p2[1], p3[1], p1[1]], 'k-')
        
        # 递归调用绘制三个子三角形
        sierpinski(ax, p1, mid_1_2, mid_1_3, depth-1)
        sierpinski(ax, mid_1_2, p2, mid_2_3, depth-1)
        sierpinski(ax, mid_1_3, mid_2_3, p3, depth-1)

# 创建一个新的图形和坐标轴
fig, ax = plt.subplots()

# 初始三角形的三个顶点
p1 = np.array([0, 0])
p2 = np.array([1, 0])
p3 = np.array([0.5, np.sqrt(3)/2])

# 调用绘制函数
sierpinski(ax, p1, p2, p3, 5)

# 设置图形的范围
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)

# 隐藏坐标轴
ax.axis('off')

# 显示图形
plt.show()
```
